```python
# Measure of central tendency

a = pd.DataFrame({'Age':[10,15,14,12,7,4,21,22]})
a
```

```python
# mean = sum of values / number of values
a['Age'].mean()
```

```python
c = 0
s = 0

for i in a['Age']:
  c = c + i
  s = s + 1

print(f'Mean of Age column : {c / s}')
```

```python
a['Age'].median()
```

```python
a['Age'].mode()
```

```python
a1 = []
b = []

for i in a['Age']:
  if i not in a1:
    a1.append(i)
  else:
    b.append(i)
print(b)
```

```python
# variance and standard_deviation

df = pd.DataFrame({'kohli':[34,30,32,28,36,33] ,
                   'rohith':[80,49,0,4,18,52]})
df
```

```python
print(f'Mean for Kohli : {df["kohli"].mean()}')
print(f'Mean for rohith: {df["rohith"].mean()}')
```

```python
print(f'Var for Kohli : {df["kohli"].var()}')
print(f'Var for rohith: {df["rohith"].var()}')
```

```python
print(f'std for Kohli : {df["kohli"].std()}')
print(f'std for rohith: {df["rohith"].std()}')
```

```python
# understanding the std in depth
df
```

```python
plt.figure(figsize=(5,3))
df['kohli'].plot(kind='kde')
plt.show()
```

```python
# giving the std to z_score = xi - mean / std
s = []
for i in df['kohli']:
  s.append((i - df['kohli'].mean()) / (df['kohli'].std()))
df['kohli_scaled'] = s
```

```python
df
```

```python
plt.figure(figsize=(5,3))
df['kohli_scaled'].plot(kind='kde')
plt.show()
```
